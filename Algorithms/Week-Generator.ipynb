```python
import pandas as pd
import re
import datetime
```

## Get busy weeks logic

```python
courses = pd.read_csv("CourseList - Sheet1.csv")

course_arr = []
for index, i in enumerate(courses["(Delivarable, Date, Diffuculty (in hours), Weight)"]):
    due_dates = re.findall("\((.+?%)\)", i)
    for arr in due_dates:
        name, date, hours, weights = map(lambda x: x.strip(), arr.split(","))
        year, month, day = map(lambda x: int(x), date.split("-"))
        
        # Return values
        hours = int(re.findall("\d+", hours)[0])
        date_week = datetime.datetime(year, month, day).isocalendar()[1]
        date_day = (datetime.datetime(year, month, day) - datetime.datetime(2020, 1, 1)).days + 1
        weight = int(re.findall("\d+", weights)[0])
        
        course_name = courses["Course"][index]
        
        date_map = {
            "name":  "{}: {}".format(course_name, name),
            "date": date,
            "date_week": date_week,
            "date_day": date_day,
            "hours": hours,
            "weight": weight,
        }
        course_arr.append(date_map)
```

```python
sorted_by_weight = pd.DataFrame(course_arr).sort_values(by=['date_day'], ascending=False)
```

```python
sorted_by_weight
```

```python
todo_list = [[] for i in range(0, 119)]
time_left = [6 for i in range(0, 119)]
for index, row in sorted_by_weight.iterrows():
    name = row["name"]
    day = row["date_day"] - 1
    hours = row["hours"]
    
    while hours > 0:
        
        # Seatch for a free day
        while (time_left[day] - 2) < 0:
            day = day - 1
        
        # Update hours left on that day
        time_left[day] = time_left[day] - 2
        # Add TODO list item
        todo_list[day].append("{}: {}".format(name, "2 Hours"))
        
        # Update all local variables
        hours = hours - 2
        day = day - 1
        
```
